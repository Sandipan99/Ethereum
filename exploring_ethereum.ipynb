```python
def value(t_x_t,t_x_f):
    val = {}
    for key in t_x_t:
        val[key] = t_x_t[key]
    for key in t_x_f:
        if key in val:
            val[key]-=t_x_f[key]
    return val

t_x = value(t_t,t_f)
```

```python
import matplotlib.pyplot as plt
import numpy as np
def CDF(lst):
    counter = {}
    for obj in lst:
        if obj not in counter:
            counter[obj] = 1
        else:
            counter[obj] +=1        
    x = []
    y = []
    cnt = 0
    for i,j in sorted(counter.items(),key = lambda x:x[0]):
        if i>=0:
            x.append(i)
            y.append(j)
        else:
            cnt+=1
    print(cnt)
    x = np.array(x)
    y = np.cumsum(np.array(y))
        
    plt.plot(x,y)
    plt.xscale('log')
    plt.yscale('log')
CDF(to_plot)   
```
